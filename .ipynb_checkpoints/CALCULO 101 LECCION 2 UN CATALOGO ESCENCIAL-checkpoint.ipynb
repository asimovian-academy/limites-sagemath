### Modelado matemático

Un *modelo matemático* es na descripción en lenguaje abstracto de una fenómeno del mundo físico. Generalmente se busca hacer predicciones a partir de dicho modelo.

#### Modelo lineal

$$ y = f(x) $$

#### Ejemplo

* Conforme el aire seco se mueve hacia arriba, se expande y enfría. Si la temperatura del suelo es `20°C` y la temperatura a una altura de `1km` es `10°C`, exprese la temperatura `T` como función de la altura `h`, suponiendo que el mode lo lineal es apropiado. 
* Trace la gráfica de la función anterior. ¿Qué es lo que representa la pendiente?
* ¿Cuál es la temperatura a una altura de `2.5kms`?

```python
# Calculamos la pendiente
h0,T0 = [0,20] #datos iniciales
h1,T1 = [1,10] #datos terminales

m = (T1-T0)/(h1-h0)

#definimos la función
h = var("h")
T(h) = T0 + m*(h-h0)
show(T)
```

```python
#de manera equivalente
T(h) = T1 + m*(h-h1)
show(T)
```

```python
#trazamos la función
plot(T, (h,0,3))
```

```python
#calculamos la temperatura para 2.5 kms
print(T(2.5))
```

### Polinomios

$$ P(x) = a_n x^n + ... + a_2 x^2 +a_1 x +a_0 $$

Son especialmente utiles para aproximar funciones, cuya expresión es desconocida. Revisemos nuevamente nuestro ejemplo sobre la población mundial

```python
poblacion = [
    [1900, 1650],
    [1910, 1750],
    [1920, 1860],
    [1930, 2070],
    [1940, 2300],
    [1950, 2560],
    [1960, 3040],
    [1970, 3710],
    [1980, 4450],
    [1990, 5280],
    [2000, 6080],
    [2010, 6870]
            ]
```

```python
var('a,b,c')
mymodel(x) = a*x^2 + b*x + c 
myfit = find_fit(poblacion, mymodel, solution_dict=True)
myfit

p = mymodel(
    a=myfit[a],
    b=myfit[b],
    c=myfit[c]
    )

show(p)

points(poblacion, color='purple') + plot(
  p,(x,1900,2010,),
    color='red'
  )

```

#### Funciones racionales

Son cocientes de polinomios. La complejidad de estas funciones deriva de las singularidades que presenta en su dominio. Por ejemplo, analicemos la función
$$ f(x) = \dfrac{2x^4-x^2+1}{x^2-4} $$

```python
x = var("x")
f(x) = (2*x^4-x^2+1)/(x^2-4)
show(f)
plot(f)
```

Para un mejor analisis, consideraremos los ceros del denominador.

```python
solve(x^2-4, x)
```

Graficamos en un intervalo amplio que incluya a $x=0$ y $x=1$.

```python
plot(f, (x,-3,3), gridlines=True, ymin=-100, ymax=100)
```

Más interesante es analizar el comportamiento asintótico. Primero simplifiquemos la función.

```python
show(f.partial_fraction(x))
```

Al compararla con la parte entera de la división larga de polinomios, vemos que cuando $x \to \pm \infty$, ambas funciones empiezan a converger. 

```python
R = 10
M = 100
grafica = plot(f, (x,-R,R), gridlines=True, ymin=-M, ymax=M)
grafica += plot(2*x^2, (x,-R,R), color = "red",
                gridlines=True, ymin=-M, ymax=M)
show(grafica)
```

#### Funciones trigonométricas

Las funciones trigonométricas básicas son el seno, el coseno y su cociente, la tangente. Las funciones coseno y seno son básicas cuando se modelan ondas mecánica:

$$ y(t) = A\sin(2\pi f t +\phi) = A\sin(\omega t + \phi)$$

donde 
* $A$ es la amplitud
* $f$ es la frecuencia ordinaria (oscilaciones por segundo)
* $\omega$ la frecuencia angular (radianes por segundo) 
* $\phi$ la fase

Esta onda esta desplazada en $\phi/\omega$ segundos. Si la cantidad es negativa indica un retardo, y si es positiva, un avance.

![IM10102](IM10102.gif)

```python
t = var("t")
A = 2
f = 10
w = 2*pi*f
phi = pi/4
y(t) = A*sin(w*t+phi)
plot(y, (t,0-phi/w,1-phi/w))
```

#### Funciones exponencial y logarítmica

Finalmente, ilustraremos algunas aplicaciones de estas funciones. Regresemos al ejemplo de la población mundial. 


```python
poblacion = [
    [1900, 1650],
    [1910, 1750],
    [1920, 1860],
    [1930, 2070],
    [1940, 2300],
    [1950, 2560],
    [1960, 3040],
    [1970, 3710],
    [1980, 4450],
    [1990, 5280],
    [2000, 6080],
    [2010, 6870]
            ]
```

Resolviendo la *ecuación diferencial*

$$ D_t P(t) = k P(t) , P(t_0) = P_0$$

obtenemos la función $P(t) = P_0 e^{k(t-t_0)} $, donde $P(t)$ es la población en función del tiempo y $P_0$ es la población inicial al tiempo $t_0$.

En nuestro ejemplo $t_0 = 1900, P_0= 1650$. 

Sin embargo, queda pendiente obtener $k$. Para esto, evaluamos en otro tiempo, digamos $ t= 1970 $, de lo cual obtenemos:

$$ 3710 = 1650e^{k(1970-1910)} $$

Utilizando logarítmos, resolvermos para $k$

```python
t0, p0 = [1900, 1650]
t1, p1 = [1970, 3710]

k = (1/(t1-t0))*ln(p1/p0)
print(k,"=", k.n())
```

Ahora, sustituimos en nuestro modelo y graficamos los datos contra las proyecciones

```python
t = var("t")
P(t) = 1650*e^(k*(t-t0))
show(P)
```

```python
grafica = points(poblacion, color='purple')
grafica += plot(P, (t,1900,2010))
show(grafica)
```

### Transformaciones

Consideremos alguna función arbitraría, digamos 
$$ f(x) = e^{-x^2}\sin(x) $$

Considere la siguiente transformación y deduzca que efecto tienen en la gráfica para diferentes valores de $h,k$:
$$g(x) = f(x-h)+k$$

```python
x = var("x")
f(x) = exp(-x^2)*sin(x)

radio = 5
plot(f, (x,-radio, radio))
```

```python
h,k = [2,3]
g(x) = f(x-h)+k

grafica = plot(g, (x, h-radio, k+radio), gridlines=True, color="red")
grafica += plot(f, (x,-radio, radio))
show(grafica)
```

Ahora, considermos que pasa al multiplicar la función por alguna constante:

$$ d(x) = c*f(x) $$

¿Qué sucede en los siguientes casos?
* $c\leq -1$
* $-1 < c < 0$ 
* $0 < c < 1$
* $1 \leq c $

```python
h,k = [2,3]
g(x) = f(x)

c = .5
d(x) = c*g(x)

grafica = plot(d, (x, h-radio, k+radio), gridlines=True, color="red")
grafica += plot(f, (x,-radio, radio))
show(grafica)
```

### Operaciones entre funciones

Las funciones tambien pueden sumarse, restarse, multiplicarse y dividirse. 
Consideremos algunos ejemplos:
    
$$ f(x) =\sin(x), g(x) = e^{-x}$$

```python
x = var("x")
f(x) = sin(x)
g(x) = exp(-x)
```

```python
#### suma

h(x) = f(x)+g(x)
show(h)

intervalo = (x,0,5)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red") + \
    plot(h, intervalo, color="green")
show(grafica)
```

```python
#### resta

h(x) = f(x)-g(x)
show(h)

intervalo = (x,0,5)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red") + \
    plot(h, intervalo, color="green")
show(grafica)
```

```python
#### multiplicacion

h(x) = f(x)*g(x)
show(h)

intervalo = (x,0,10)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red") + \
    plot(h, intervalo, color="green")
show(grafica)
```

```python
#### division

h(x) = f(x)/g(x)
show(h)

intervalo = (x,0,pi)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red") + \
    plot(h, intervalo, color="green")
show(grafica)
```

```python
intervalo = (x,0,2*pi)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red") + \
    plot(h, intervalo, color="green")
show(grafica)
```

```python
intervalo = (x,0,3*pi)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red") + \
    plot(h, intervalo, color="green")
show(grafica)
```

Por último, consideraremos la composición de funciones que se define como 
$$ (f\circ g)(x) = f(g(x))$$

Consideremos ahora
$$ f(x)= \cos(x), g(x) = \dfrac{1}{x}$$

```python
f(x) = cos(x)
g(x) = 1/x

h(x) = f(g(x))
show(h)

intervalo = (x,-2*pi, 2*pi)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red", ymin=-2, ymax=2) + \
    plot(h, intervalo, color="green")
show(grafica)
```

Observe que el orden de la composición nos da dos resultados diferentes en general

```python
f(x) = cos(x)
g(x) = 1/x

h(x) = g(f(x))
show(h)

intervalo = (x,-2*pi, 2*pi)
grafica = plot(f, intervalo) + \
    plot(g, intervalo, color="red", ymin=-2, ymax=2) + \
    plot(h, intervalo, color="green")
show(grafica)
```
